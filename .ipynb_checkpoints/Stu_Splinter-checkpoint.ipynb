# Windows Users

Windows users will need to store the `chromedriver.exe` file in the same location as this _.ipynb_ file, or they will need to provide the path to the executable.

```python
executable_path = {'executable_path': 'chromedriver.exe'}
browser = Browser('chrome', **executable_path, headless=False)
```

### Define URL to scrape and inform the browser to visit the page

```python
url = 'http://books.toscrape.com/'
browser.visit(url)
```

### Iterate over multiple pages and scrape content from each

```python
for x in range(50):
    
    # Create HTML object
    html = browser.html
    
    
    
    ########## STUDENT CODE HERE ##########
    # Parse HTML with Beautiful Soup
    soup = BeautifulSoup(html, 'html.parser')
    # Retrieve all elements that contain book information
    articles = soup.find_all('article', class_ = 'product_pod')
    ########## STUDENT CODE END ##########
    
    
    # Iterate through each book
    for article in articles:
        
        ########## STUDENT CODE HERE ##########
        # Use Beautiful Soup's find() method to navigate and retrieve title and URL
        link = article.find('h3').find('a')
        title = link['title']
        url = link['href']

        
        ########## STUDENT CODE END ##########
        
        print('-----------')
        print(title)
        print('http://books.toscrape.com/' + url)
        
    

    
#     try:
#         ########## STUDENT CODE HERE ##########
#         # Click the 'Next' button on each page
        
#         ########## STUDENT CODE END ##########
#     except:
#         print("Scraping Complete")
```
